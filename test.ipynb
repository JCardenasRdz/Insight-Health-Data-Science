# Simulate some data

```python
# boilter plate code
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# CEST for simulation
from qMRI.cest import lorentzian_sim

a = [0.9, 0.7, 0.5, 0.5]
b = [2,1,1,1]
c = [0,1,4.2,-5.6]
x = np.linspace(-20,20,101)

yobserved = lorentzian_sim(x,a,b,c)
yobserved = yobserved + np.random.normal(0,.05,yobserved.shape)
```

# Perform Non-Linear Fitting

```python
from qMRI.cest import lorentzian_fit
xpredicted, Shat, xMatrix = lorentzian_fit(x,yobserved,c, repetitions = 10)
```

## Plot Results

```python
plt.plot(x,1-Shat,x,1-yobserved,'or'); plt.legend(('Predicted','Observed'))
plt.xlabel('Saturation Offset (ppm)')
plt.ylabel('Signal(Mz/Mo)')
```
